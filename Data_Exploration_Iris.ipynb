```python
##sqrt((x - y)' * Q * (x - y))
def mahalanobis_dist(vec_A,vec_B,M= None):
  mu = vec_A - vec_B
  if M is None:
      M = np.cov(vec_A.T)
      inv = np.linalg.inv(M)
      dp = np.dot(mu,inv)
      mah_dis = np.dot(dp,mu.T)
  return np.sqrt(mah_dis.diagonal())  


```
